```python
from datetime import datetime

df['fec_not2']=pd.to_datetime(df['fec_not'],format="%d/%m/%Y")
df['fecha_contagio2']=pd.to_datetime(df['fecha_contagio'],format="%d/%m/%Y")
df['fec_hos_2']=pd.to_datetime(df['fec_hos_'],format="%d/%m/%Y")
df['fecha_defuncion2']=pd.to_datetime(df['fecha_defuncion'],format="%d/%m/%Y")
df['fecha_nacimiento2']=pd.to_datetime(df['fecha_nacimiento'],format="%d/%m/%Y")
```

```python
import pyproj
P=pyproj.Proj(proj='utm', zone=18, ellps='WGS84', preserve_units=True)
```

```python
df['x'],df['y']=P(df['longitud'],df['latitud'])
df
```

coordenadas 7.063042872866057, -73.08617624275931

```python
import math
lat_alcaldia=7.063042872866057
long_alcaldia=-73.08617624275931
x_alcaldia,y_alcaldia=P(long_alcaldia,lat_alcaldia)
x_alcaldia,y_alcaldia
df


def calcular_distancia(x1,y1,x2,y2):
  distancia=(pow((x1-x2),2)+pow((y1-y2),2))**1/2
  return distancia
```

```python
df['distancia']=calcular_distancia(df['x'],df['y'],x_alcaldia,y_alcaldia)
df
```

```python
datosf=df
datosf=datosf.select_dtypes(exclude=["object","datetime64"])
datosf= datosf.drop("año", axis=1)
datosf= datosf.drop("cod_pais_o", axis=1)
datosf= datosf.drop("condicion_final", axis=1)
datosf= datosf.drop("fecha_defuncion", axis=1)
datosf= datosf.drop("certif_defuncion", axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sb 
colormap= plt.cm.viridis
plt.figure(figsize=(24,24))
plt.title("Pearson correlation of features", y= 1.05, size=15)
sb.heatmap(datosf.astype(float).corr(),linewidths=1, vmax=1.0, square=True, cmap=colormap, linecolor="black", annot=True)
```

```python
ab=100 #ancho de banda
res=10 #resolucion
varKernel=3 #funcion de kernel
#1 Uniforme
#2 Triangular
#3 Gaussiana
#4 Epanechnikov
#5 Varianza mínima
maxx=max(df["x"])+ab
maxy=max(df["y"])+ab
minx=min(df["x"])-ab
miny=min(df["y"])-ab
import numpy as np

filas=math.ceil((maxy-miny)/res)
print(filas)

columnas=math.ceil((maxx-minx)/res)
print(columnas)
matrizHM=np.zeros((filas,columnas), dtype=float)

for i in range(len(df)):
    y=math.floor((df.iloc[i]['y']-miny)/res)
    x=math.floor((df.iloc[i]['x']-minx)/res)
    matrizHM[y][x]+=1
    
sb.set(rc = {'figure.figsize':(30,30)})
ax = sb.heatmap(matrizHM)
```

```python
import numpy as np
matrizAcumuladora=np.zeros((filas,columnas), dtype=float)
var=ab//res
var, filas, columnas
```

```python
def FuncionesKernel(dist):
  if(varKernel==1):
    y=0.5
  elif(varKernel==2):
    y=1-(abs((dist)/(ab)))
  elif(varKernel==3):
    y=(np.exp(-(((dist)/(ab))**2/(2)))/(2*math.pi)**0.5)
  elif(varKernel==4):
    y=(3/4)*(1-(dist/ab)**2)
  elif(varKernel==5):
    y=3/8(3-5*(dist/ab)**2)
  return y
  
```

```python
for i in range(filas):
  for j in range(columnas):
    for k in range(max(i-var,0), min(i+var+1,filas)):
      for l in range(max(j-var,0), min(j+var+1,columnas)):
        distancia = res * (((j - l)**2 + (i-k)**2)**0.5)
        if(distancia<=ab):
          
         matrizAcumuladora[i][j]+=matrizHM[k][l]*FuncionesKernel(distancia)

matrizAcumuladora
sb.set(rc = {'figure.figsize':(30,30)})
bx = sb.heatmap(matrizAcumuladora)
```

```python

import matplotlib.pyplot as plt

fig = plt.figure(figsize=(25, 18))

ax = fig.add_subplot(111)
ax.set_title('Mapa de calor')
plt.imshow(matrizAcumuladora) # <----- Matriz
ax.set_aspect('equal')

cax = fig.add_axes([0.12, 0.1, 0.78, 0.8])
cax.get_xaxis().set_visible(False)
cax.get_yaxis().set_visible(False)
cax.patch.set_alpha(0)
cax.set_frame_on(False)
plt.colorbar(orientation='vertical', pad=0.2)
plt.show()
```
